Drop duplicates

```python
data = [
    (1, "Alice", "HR", 5000, "2024-01-01"),
    (1, "Alice", "HR", 5000, "2024-01-01"),  # exact duplicate
    (2, "Bob", "IT", 6000, "2024-01-02"),
    (2, "Bob", "IT", 6500, "2024-01-05"),   # duplicate id, different salary
    (3, "Charlie", "IT", 7000, "2024-01-03"),
    (3, "Charlie", "IT", 7000, "2024-01-03"),
    (4, "Charliek", "IT", 7000, "2024-01-03"),
    (5, "Charliet", "IT", 8000, "2024-01-03") # exact duplicate
]

columns = ["id", "name", "dept", "salary", "date"]

df = spark.createDataFrame(data, columns)
df.show()
```

1️⃣ Drop exact duplicate rows

```python
df1 = df.dropDuplicates()
df1.show()
```

Drop duplicates based on single column

```python
df2 = df.dropDuplicates(["ID"])
df2.show()
```

```python
df1 = df.dropDuplicates(["salary"])
df1.show()
```

Drop duplicates based on multiple columns

```python
df1 = df.dropDuplicates(["Dept", "ID"])
df1.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

w = Window.partitionBy("id").orderBy(desc("date"))
row_df = df.withColumn("row", row_number().over(w)).filter("row = 1")
row_df.show()

```

Keep record with max salary

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

w = Window.partitionBy("id").orderBy(desc("salary"))

df1 = df.withColumn("row", row_number().over(w)).filter("row = 1").drop("row")
display(df1)
```
